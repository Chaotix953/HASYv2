```python
class CustomDataset(Dataset):
    def __init__(self, dataframe):
        self.dataframe = dataframe
        self.class_to_idx = {cls: i for i, cls in enumerate(sorted(df['symbol_id'].unique()))}
        self.idx_to_class = {idx: cls for cls, idx in self.class_to_idx.items()}

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.loc[idx, 'path']
        symbol_id = self.dataframe.loc[idx, 'symbol_id']
        image = torchvision.io.read_image(img_path.replace('../../', ''))
        image = transforms.Grayscale(num_output_channels=1)(image) 
        image = image / 254
        image = image.squeeze(dim=2)
        label = self.class_to_idx[symbol_id]
        return image, label
```

```python
import torch.nn as nn

class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=16, kernel_size=5, stride=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=5, stride=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, stride=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.fc1 = nn.Linear(64 * 4 * 4, 512) 
        self.fc2 = nn.Linear(512, 369)        

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size(0), -1)
        x = self.fc1(x)   
        output = self.fc2(x)  
        return output

    def forward_(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size(0), -1)
        output = self.fc1(x)  
        return output
```

```python
def load_dataset(path, batch_size, shuffle=True):
    df = pd.read_csv(path)
    dataset = CustomDataset(df)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle, pin_memory=True)
    return dataloader
```

```python
train_loader = load_dataset("classification-task/fold-1/train.csv", 16)
test_loader = load_dataset("classification-task/fold-1/test.csv", 16)
```

```python
model = ConvNet().to(device, non_blocking=True)
criterion = nn.CrossEntropyLoss().to(device, non_blocking=True)
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
print(train_loader.dataset[0][0].shape)
```

```python
from torchmetrics import Accuracy

num_epochs = 8
accuracy_metric = Accuracy(task='multiclass', num_classes=369).to(device, non_blocking=True)

model.train()
for epoch in range(num_epochs):
    running_loss = 0.0
    accuracy_metric.reset()  # Réinitialisez la métrique à chaque époque

    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}")
    for images, labels in pbar:
        images, labels = images.to(device, non_blocking=True), labels.to(device, non_blocking=True)
        optimizer.zero_grad()
        outputs = model.forward(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        accuracy_metric.update(outputs, labels)  # Mise à jour de la métrique d'exactitude

        # Calcul de la perte moyenne et de l'exactitude en temps réel
        pbar.set_postfix(loss=running_loss / len(train_loader.dataset), accuracy=accuracy_metric.compute().item())

    epoch_loss = running_loss / len(train_loader)
    epoch_accuracy = accuracy_metric.compute()  # Obtenez la précision moyenne pour l'époque
    print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")

print('Finished Training')
```

```python
running_loss = 0.0
correct = 0
total = 0

with torch.no_grad():
    for images, labels in tqdm(test_loader, desc="Evaluation", leave=False):
        images, labels = images.to(device), labels.to(device)
        outputs = model.forward(images)
        loss = criterion(outputs, labels)
        running_loss += loss.item()
        predicted = torch.argmax(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

epoch_loss = running_loss / len(test_loader)
epoch_accuracy = correct / total * 100
print(f"Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.2f}%")
```

```python
model = ConvNet()
model.eval()

# Fonction pour extraire les features
def get_features(data_loader, model):
    model = model.to(device)
    features = []
    labels = []
    with torch.no_grad():
        for images, targets in data_loader:
            images = images.to(device)
            output_features = model.forward_(images)
            features.append(output_features.to('cpu').numpy())
            labels.append(targets.numpy())
    return np.concatenate(features), np.concatenate(labels)
    
features, labels = get_features(test_loader, model)
tsne = TSNE(n_components=2, random_state=0)
X_2d = tsne.fit_transform(features)

plt.figure(figsize=(6, 5))
for i in range(0, 369):
    mask = labels == i
    plt.scatter(X_2d[mask, 0], X_2d[mask, 1], label=i)
plt.legend()
plt.title('t-SNE projection of the MNIST test set')
plt.show()
```
